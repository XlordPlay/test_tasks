```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#take ur path
data = pd.read_csv('/home/xlordplay/test_tasks/task_type_3/data.csv')

data.head()
```

```python

data.info()


data.isnull().sum()
```

```python

data['date'] = pd.to_datetime(data['date'], errors='coerce')
data = data.dropna(subset=['date'])  # Удаляем строки с некорректными датами

# Установим индекс по дате и отсортируем данные
data.set_index('date', inplace=True)
data = data.sort_index()

# Визуализация цены закрытия
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['close'], label='Цена закрытия')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.title('Тренд цены закрытия криптовалюты')
plt.legend()
plt.show()
```

```python
"""
Визуализация распределения ключевых показателей, таких как цена закрытия, открытия, максимальная и минимальная цена, помогает:

Понять форму распределения данных. Например, если данные о цене закрытия сильно смещены в одну сторону, 
это может свидетельствовать о тренде или асимметричном поведении рынка.

Проверить, есть ли выбросы или необычно высокие/низкие значения. Это важно, потому что выбросы могут негативно повлиять на точность модели, 
если не обработать их должным образом.

Гистограммы показывают, как часто встречаются те или иные значения. 
Например, если гистограмма для колонки close сильно скошена влево или вправо, это может указывать на преобладание трендов
на повышение или понижение цены.
"""
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
data['close'].plot(kind='hist', bins=50, ax=axes[0, 0], color='skyblue', edgecolor='black', title='Распределение цены закрытия')
data['open'].plot(kind='hist', bins=50, ax=axes[0, 1], color='lightgreen', edgecolor='black', title='Распределение цены открытия')
data['high'].plot(kind='hist', bins=50, ax=axes[1, 0], color='salmon', edgecolor='black', title='Распределение максимальной цены')
data['low'].plot(kind='hist', bins=50, ax=axes[1, 1], color='orange', edgecolor='black', title='Распределение минимальной цены')
plt.tight_layout()
plt.show()
```

```python
#Корреляционная матрица и тепловая карта
"""
Корреляция помогает понять, как разные признаки (например, open, close, high, low, volume) взаимосвязаны между собой:

Высокая корреляция между переменными, например, между high и low, может говорить о том, 
что они двигаются в одном направлении, и мы можем использовать один из них в качестве индикатора для предсказания другого.
Корреляция с ценой закрытия (close) позволяет определить, какие факторы (например, объём торгов или спред) влияют на конечную цену.

Если показатели сильно коррелируют друг с другом, можно снизить размерность данных, исключив одну из переменных. 
Это также помогает выявить признаки, которые наиболее влияют на цену криптовалюты.
"""
plt.figure(figsize=(10, 6))
correlation_matrix = data[['open', 'high', 'low', 'close', 'volume', 'market', 'spread']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Корреляция между показателями")
plt.show()
```

```python
# Средняя цена закрытия за каждый месяц
monthly_avg = data['close'].resample('M').mean()
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg, color='purple', label='Средняя месячная цена')
plt.xlabel('Дата')
plt.ylabel('Средняя цена')
plt.title('Средняя цена закрытия по месяцам')
plt.legend()
plt.show()

"""
Для анализа сезонности мы можем рассмотреть поведение цены на уровне месяца или недели. 
Например, можно найти среднюю цену за каждый месяц или день недели.
"""

# Средняя цена закрытия по дням недели (0 - Понедельник, 6 - Воскресенье)
data['weekday'] = data.index.weekday
weekday_avg = data.groupby('weekday')['close'].mean()
plt.figure(figsize=(10, 5))
sns.barplot(x=weekday_avg.index, y=weekday_avg.values, palette='viridis')
plt.xlabel('День недели')
plt.ylabel('Средняя цена')
plt.title('Средняя цена закрытия по дням недели')
plt.show()
```

```python
# Рассчитаем дневные процентные изменения цены
data['daily_change'] = data['close'].pct_change() * 100

""" 
Построим график волатильности
Волатильность — важный показатель для криптовалют. 
Мы можем рассчитать процентное изменение цены по дням и визуализировать его, чтобы увидеть, как часто происходят резкие колебания.
"""
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['daily_change'], color='red', label='Дневное изменение цены (%)')
plt.axhline(0, color='black', linestyle='--', lw=1)
plt.xlabel('Дата')
plt.ylabel('Изменение (%)')
plt.title('Дневная волатильность цены')
plt.legend()
plt.show()
```

```python
""" 
Построим график для определения возможных аномалий — резких скачков или падений цены, 
которые могут быть вызваны внешними событиями.
"""
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['close'], label='Цена закрытия', color='blue')
plt.scatter(data.index[data['daily_change'].abs() > 10], data['close'][data['daily_change'].abs() > 10], 
            color='red', label='Аномальные изменения', s=50)
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.title('Аномальные изменения цены закрытия')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

#Нормализация данных
scaler = MinMaxScaler(feature_range=(0, 1))
data['close'] = scaler.fit_transform(data[['close']])


train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size]
test_data = data.iloc[train_size:]

print(f"Размер обучающего набора: {len(train_data)}")
print(f"Размер тестового набора: {len(test_data)}")
```

```python
import numpy as np

def create_dataset(dataset, time_step=1):
    X, Y = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i + time_step), 0]
        X.append(a)
        Y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(Y)

#Определим размер окна (например, 60 дней)
time_step = 60
train_values = train_data[['close']].values
test_values = test_data[['close']].values

X_train, y_train = create_dataset(train_values, time_step)
X_test, y_test = create_dataset(test_values, time_step)

#Изменим форму данных для модели LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import numpy as np

# Нормализация данных (преобразуем данные в диапазон [0, 1])
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

# Преобразуем X_train и y_train в двумерный массив для масштабирования
X_train_scaled = np.reshape(X_train, (X_train.shape[0] * X_train.shape[1], X_train.shape[2]))
y_train_scaled = np.reshape(y_train, (-1, 1))

# Масштабируем данные
X_train_scaled = scaler_X.fit_transform(X_train_scaled)
y_train_scaled = scaler_y.fit_transform(y_train_scaled)

# Восстановим данные в исходную форму (три измерения)
X_train_scaled = np.reshape(X_train_scaled, (X_train.shape[0], X_train.shape[1], X_train.shape[2]))
y_train_scaled = np.reshape(y_train_scaled, (y_train.shape[0], 1))

# Разделение на тренировочные и валидационные данные
X_train, X_val, y_train, y_val = train_test_split(X_train_scaled, y_train_scaled, test_size=0.2, shuffle=False)

# Построение модели
model = Sequential()

# Первый LSTM слой с регуляризацией
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2]), kernel_regularizer=l2(0.01)))

# Второй LSTM слой с регуляризацией
model.add(LSTM(50, return_sequences=False, kernel_regularizer=l2(0.01)))

# Полносвязный слой
model.add(Dense(25))

# Выходной слой
model.add(Dense(1))

# Компиляция модели
model.compile(optimizer='adam', loss='mean_squared_error')

# Обучение модели с добавлением валидационной выборки
model.fit(X_train, y_train, 
          batch_size=16, 
          epochs=50,  
          validation_data=(X_val, y_val))  # Добавляем валидацию



```

```python
# Прогнозирование на тестовых данных
predictions = model.predict(X_test)

# Обратное масштабирование предсказаний и истинных значений, чтобы вернуть их к исходной шкале
predictions = scaler_y.inverse_transform(predictions)  # scaler_y для y (target)

# Обратное масштабирование y_test
y_test_scaled = scaler_y.inverse_transform(y_test.reshape(-1, 1))  # scaler_y для y (target)

# RMSE
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test_scaled, predictions))
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Построение графика
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data.index[train_size + time_step + 1:], y_test_scaled, label='Фактические значения')
plt.plot(data.index[train_size + time_step + 1:], predictions, label='Прогнозируемые значения')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.title('Сравнение фактических и прогнозируемых значений')
plt.legend()
plt.show()

```

```python
#Значение RMSE показывает, 
#что модель +- точно предсказывает значения (средняя ошибка около 0.08 в исходной шкале цены).
```

```python
# Предсказание на 10 дней вперед
last_data = test_data['close'].values[-time_step:]
temp_input = last_data.reshape(1, -1).tolist()[0]

predictions_future = []

# Прогнозируем на 10 дней вперед
for i in range(10):
    if len(temp_input) > time_step:
        temp_input = temp_input[1:]
    input_data = np.array(temp_input).reshape(1, time_step, 1)
    pred = model.predict(input_data, verbose=0)
    temp_input.append(pred[0][0])
    predictions_future.append(pred[0][0])

# Обратное масштабирование предсказанных значений
predictions_future = scaler_y.inverse_transform(np.array(predictions_future).reshape(-1, 1))

# Создадим даты для прогноза
future_dates = pd.date_range(data.index[-1] + pd.Timedelta(days=1), periods=10)

# Построим график с прогнозом на 10 дней вперед
plt.figure(figsize=(12, 6))
plt.plot(data.index, scaler_y.inverse_transform(data[['close']]), label='Фактические значения')  # Используем scaler_y для обратного масштабирования
plt.plot(future_dates, predictions_future, label='Прогноз на 10 дней вперед', linestyle='--')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.title('Прогноз курса криптовалюты на 10 дней вперед')
plt.legend()
plt.show()

```
